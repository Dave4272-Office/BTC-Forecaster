# Technical Indicators
Adding technical indicator values

### Importing Libraries and Data

```python
import pandas as pd
import _pickle as pickle
import numpy as np
import matplotlib as plt
plt.style.use('bmh')
```

```python
df = pd.read_csv('BTC-USD.csv')

# Viewing the DF
df
```

### Datetime Conversion

```python
# Datetime conversion
df['Date'] = pd.to_datetime(df.Date)

# Setting the index
df.set_index('Date', inplace=True)
```

## Charting Values

```python
# Viewing the Chart of Closing Values
df['Close'].plot('line', figsize=(15,7), title='BTC Closing Prices');
```

## Creating the Technical Indicators

```python
np.mean(df['Close'][-10:-1])
```

```python
def SMA(periods, col='Close'):
    """
    Calculating the Simple Moving Average for the past n days
    """
    return round(np.mean(df[col][-periods:-1]), 2)
```

```python
df['SMA'] = df.apply(lambda row: SMA(df, periods=10))
```
